##### Scraping Detailed Beer Recipe Page
Here we demonstrate how to use `scrape_recipes_details` to scrape each beer recipe detailed page. Given a list of beer recipe's url, our function attempts to make scape  all of the beer recipes' details such as ingredients used, ratings and comments left behind. Snippet below showcase the scraping done for just 10 recipes. Be mindful that the entire dataset has **over 100k** recipes and thus will take **over 16 hours** if attempted to run within this notebook. Instructions on how to effectively run and scrape all the detailed data can be found inside `scripts/parallel-reviews.py` and `scripts/parallel-ingredients.py`.

```python
# Scraping all of the beer recipes' details such as ingredients used, ratings and comments left behind
# and write it out into a two CSV files, recipe_ingredient_10_test.csv and 
# recipe_review_10_test.csv
beer_recipes = pd.read_csv("./data/raw/all_recipes.csv", keep_default_na=False)
title_10_urls = beer_recipes.title_url.loc[0:9]
title_10_urls
sb.scrape_recipes_details(title_10_urls, "recipe_ingredient_10_test.csv", "recipe_review_10_test.csv")
```

## Prepping Raw Data

#### Joining CSV Files
Since we are scrapping from multiple pages, we will need to end up joining the dataset in order to actual do any cleaning and modeling. Below shows how we joined two CSV files of 100,000+ beer recipes.

```python
all_general = pd.read_csv("./data/raw/all_recipes.csv", keep_default_na=False)
all_ingredients = pd.read_csv("./data/raw/brewer_ingredients_all.csv", keep_default_na=False)

result = pd.concat([all_general, all_ingredients], axis =1)

result.to_csv("./data/prepped/new_all_recipes.csv", encoding='utf-8')
```
