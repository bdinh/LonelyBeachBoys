```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectPercentile
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import Ridge
```

```python
features = pd.read_csv('data/prepped/modeling_features.csv')
labels = pd.read_csv('data/prepped/modeling_outcome.csv')
labels = labels['rating_value']
```

```python
train_features, test_features, train_outcome, test_outcome = train_test_split(
    features,
    labels,
    test_size=0.25,
    random_state=42
)
folds = KFold(n_splits=10, shuffle=True, random_state=42)
```

```python
pipeline = make_pipeline(
    MinMaxScaler(),
    SelectPercentile(),
    SVR()
)
pipeline_params = {
    "selectpercentile__percentile": [70, 80, 90],
    "svr__kernel": ['rbf'],
    "svr__C": [1, 10, 50, 70, 100],
    "svr__gamma": ["auto"]
}
model = GridSearchCV(pipeline, pipeline_params, cv=folds)
model.fit(train_features, train_outcome)
score = model.score(test_features, test_outcome)
```

```python
print("model score:", score)
print("best params:", model.best_params_)
```

```python
pipeline = make_pipeline(
    MinMaxScaler(),
    SelectPercentile(),
    Ridge()
)
pipeline_params = {
    "selectpercentile__percentile": [75, 80, 85],
    "ridge__alpha": range(5, 10),
    "ridge__fit_intercept": ["false"]
}
model = GridSearchCV(pipeline, pipeline_params, cv=folds)
model.fit(train_features, train_outcome)
score = model.score(test_features, test_outcome)
```

```python
print("model score:", score)
print("best params:", model.best_params_)
```
